```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
paths = glob.glob('/Users/wickes1/Desktop/oldscreenshots/*.png')
```

```python
data = []

for path in paths:
    print(path)
    split = path.split()
    date = split[2]
    hour, minute = split[4].split(".")[0:2]
    ampm = split[-1].split(".")[0]
    if not ampm in ['AM', 'PM']:
        continue
    sub = [date, ampm + hour.zfill(2)]
    print(ampm)
    data.append(sub)
```

```python
df = pd.DataFrame(data, columns = ['date', 'hour'])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
list(df.groupby(df.date.dt.year))
```

```python
df.groupby('date').size().plot()
```

```python
dayOfWeek={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['dayofweek'] = df['date'].dt.dayofweek
```
